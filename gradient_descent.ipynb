```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# given a function f and a starting point W(weight) ,try to find the minima or gradient descent 
# (when we will call the function, it will keep on oscillating
```

```python
def gradient_descent(func,w) :
    list_of_weights = []
    delta = 0.0001
    weight = w 
    learning_rate = 0.1
    for i in range(1000) :
        derivative = (func(weight+delta) - func(weight)) / delta 
        weight = weight - (learning_rate * derivative)
        list_of_weights.append(weight)
    return list_of_weights
```

```python
gradient_descent(lambda x : x ** 2 + 4 * x + 3,10) 
```

```python
def f(x) :
    return x ** 2 
x = np.linspace(-5,5,1000) 
dfdx = np.gradient(f(x),x) 
plt.plot(x,f(x), label = 'f(x)')
plt.plot(x,dfdx, label = 'df/dx')
plt.legend() 
plt.show()
```
